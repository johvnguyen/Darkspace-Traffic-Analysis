```python
from scapy.all import rdpcap
from pprint import pprint
```

```python
packets = rdpcap('telescopesampleanon/pcapsanon/ucsd-nt.1660435200.pcap')
```

```python
tcp_packets = [packet for packet in packets if packet.haslayer('TCP')]
udp_packets = [packet for packet in packets if packet.haslayer('UDP')]

print(f'There are {len(tcp_packets)} TCP packets')
print(f'There are {len(udp_packets)} UDP packets')
```

```python
for packet in packets:
    if packet.haslayer('IP'):  # Check if the packet has an IP layer
        ip_layer = packet['IP']
        srcs.add(ip_layer.src)
        dests.add(ip_layer.dst)

srcs = {packet['IP'].src for packet in packets if packet.haslayer('IP')}
dsts = {packet['IP'].dst for packet in packets if packet.haslayer('IP')}


print("Source IPs: ", end = "")
pprint(srcs)

print("\nDestination IPs: ", end = "")
pprint(dsts)
```

```python
timestamps = {float(packet.time) for packet in packets}

print(f'The timestamp range is: [{min(timestamps)}, {max(timestamps)}]')
print(f'The span of time this data encompases is: {max(timestamps) - min(timestamps)} sec')
```
